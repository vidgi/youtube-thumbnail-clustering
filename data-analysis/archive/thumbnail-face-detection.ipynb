# demo of simple face detection from youtube thumbnails

```python
import cv2
import matplotlib.pyplot as plt
from urllib.request import urlopen
from PIL import Image
from skimage import io

def get_detected_face (face):  
    face_img = face.copy()  
    fd = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    fr = fd.detectMultiScale(face_img)  
    for (x,y,width,height) in fr:  
        cv2.rectangle(face_img, (x,y), (x + width, y+height), (255,255,77), 5)                
    return face_img

def detectFromUrl(url):
    face1 = io.imread(url)
    result = get_detected_face(face1)
    plt.imshow(result)
    plt.show()
```

```python
thumbnails=["https://i.ytimg.com/vi/gPdUslndvVI/0.jpg",
            "https://i.ytimg.com/vi/3C66w5Z0ixs/0.jpg",
            "https://i.ytimg.com/vi/p7HGUZWq_8s/0.jpg",
            "https://i.ytimg.com/vi/uet14uf9NsE/0.jpg",
            "https://i.ytimg.com/vi/VIUo6yapDbc/0.jpg",
            "https://i.ytimg.com/vi/CezxKs9YLv0/0.jpg",
            "https://i.ytimg.com/vi/GYeKX_LPY9U/0.jpg",
            "https://i.ytimg.com/vi/9AecsACtkB4/0.jpg",
            "https://i.ytimg.com/vi/p6RatnMhDj8/0.jpg"           ]

for i in thumbnails:
    detectFromUrl(i)
```
