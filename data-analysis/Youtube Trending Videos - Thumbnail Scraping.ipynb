# Youtube Thumbnail Scraping

### Developing/testing procedure for face detection

```python
import matplotlib.pyplot as plt
import urllib.request

def saveYoutubeThumbnail(yt_id):
    url= "https://i.ytimg.com/vi/"+yt_id+"/maxresdefault.jpg"
    image_name = "images/"+yt_id+".jpg"
    urllib.request.urlretrieve(url, image_name)

# test: 
# saveYoutubeThumbnail('WMweEpGlu_U')
```

### Library Imports

```python
import numpy as np
import pandas as pd
import math

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

## Import dataset

```python
df = pd.read_csv("US_youtube_trending_data.csv")
df.head()
```

```python
# save top 1000 thumbnails
top1000 = df.drop_duplicates(subset=['video_id'])
top1000 = top1000.sort_values(by=['view_count'],ascending=False).head(1000)
```

```python
ids = top1000['video_id'].tolist()

for i in ids:
    try:
        saveYoutubeThumbnail(i)
    except:
        print('unable to process video: '+i)
```

```python
categories =  pd.DataFrame(top1000.groupby(by=["categoryId"]).count()['video_id'])
categories = categories.rename(columns={"video_id": "count"})
categories.sort_values(['count'], ascending=False)
```
